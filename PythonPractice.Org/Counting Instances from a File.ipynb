Given a .txt file that has a list of a bunch of names, count how many of each name there are in the file, and print out the results to the screen. I have a .txt file for you, if you want to use it!

Extra:

Instead of using the .txt file from above (or instead of, if you want the challenge), take this .txt file (https://www.practicepython.org/assets/Training_01.txt) and count how many of each “category” of each image there are. This text file is actually a list of files corresponding to the SUN database scene recognition database, and lists the file directory hierarchy for the images.

```python
import re

file_name = 'Training_01.txt'
countDict = {}
```

```python
with open(file_name, 'r') as open_file:
    line = open_file.readline()
    while line:
        m = re.search('/.*/',line)
        if m.group() in countDict:
            countDict[m.group()] += 1
        else:
            countDict[m.group()] = 1
        line = open_file.readline()
```

```python
for key in countDict:
    print(f"{key}: {str(countDict[key])}")
```
